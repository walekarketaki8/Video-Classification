```python
# Pakages needed
import os
import cv2
import keras
from keras.applications import InceptionV3
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import Model
import numpy as np
import random
from keras.models import load_model
from keras.utils import to_categorical
from sklearn.utils import shuffle
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import LSTM
```

```python
# Load the pretrained InceptionV3 model and restructure it to get feature vector from last pool layer
model = InceptionV3()
# re-structure the model
model.layers.pop()
model = Model(inputs=model.inputs, outputs=model.layers[-1].output)
```

```python
# extract features from each frame in video in the directory
def extract_features(vid):
    # extract features from each photo
    features = []
    labels = []
    for name in os.listdir(vid):
        # Label
        label = (os.path.join(vid,name)).split('/')[-1]
        label = label.split('_')[0]
       
        # Capture frames from Video
        vidcap = cv2.VideoCapture(os.path.join(vid,name))
        success,image = vidcap.read()
        
        while success:
            success,image = vidcap.read()
            cv2.imwrite(os.path.join(os.getcwd(),'saved.jpg'), image)
            try:
                image = load_img(os.path.join(os.getcwd(),'saved.jpg'), target_size=(299, 299))
                # convert the image pixels to a numpy array
                image = img_to_array(image)
                # reshape data for the model
                image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
                # prepare the image for the inception model
                image = keras.applications.inception_v3.preprocess_input(image)
                # get features
                feature = model.predict(image, verbose=0)
            
                features.append([feature])
                labels.append(int(label))
                print("In progress")
            except:
                pass
    return features,labels
```

```python
# extract features from all images
directory = '/home/ketakiw/deep-learning/videos/train'
features,labels = extract_features(directory)
print('Extracted Features: %d' % len(features))
```

```python
total_frames = len(features)
print(labels)
```

```python
# Convert labels into one-hot vector
labels2 = to_categorical(labels)
print(labels2)
```

```python
main_features = features
main_labels = labels2
```

```python
# Shuffle features and labels together
list_1 = main_features
list_2 = main_labels
list_1, list_2 = shuffle(list_1, list_2)
print(list_2)
```

```python
# List to numpy array
transfer_values = np.array(features)
labels_train = np.array(labels2)
```

```python
# Reshape numpy array
transfer_values = np.reshape(transfer_values,(total_frames,2048))
```

```python
len(transfer_values)
```

```python
# Creating chunks of 15 features and respective labels
frames_num=15
count = 0
joint_transfer=[]

for i in range(int(len(transfer_values)/frames_num)):

    inc = count+frames_num

    joint_transfer.append([transfer_values[count:inc],labels_train[count]])

    count =inc
```

```python
# RNN architecture
chunk_size = 2048
n_chunks = 15
rnn_size = 512

model = Sequential()

model.add(LSTM(rnn_size, input_shape=(n_chunks, chunk_size)))

model.add(Dense(1024))

model.add(Activation('relu'))

model.add(Dense(50))

model.add(Activation('sigmoid'))

model.add(Dense(4))

model.add(Activation('softmax'))

model.compile(loss='mean_squared_error', optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
# Training loop
data =[]
target=[]
epoch = 5
batchS = 4

for i in joint_transfer:
    data.append(i[0])
    target.append(np.array(i[1]))

model.fit(np.array(data), np.array(target), epochs=epoch, batch_size=batchS, verbose=1)
```

```python
# Model Save
model.save("rnn.h5", overwrite=True)
```
